```python
# getting anti-war songs from wiki!

from bs4 import BeautifulSoup
```

```python
import requests

r = requests.get('https://en.wikipedia.org/wiki/List_of_anti-war_songs')
r.status_code
```

```python
html = r.content
html
```

```python
soup = BeautifulSoup(html, 'html.parser')
```

```python
html_table = soup.find_all('table', attrs={'class': 'wikitable sortable'})
```

```python
import pandas as pd
import numpy as np
```

```python
#into df

df = pd.read_html(str((html_table)))[0]
df
```

```python
np_array = np.array(df)
```

```python
np_array.shape
```

```python
reshaped_array = np.reshape(np_array, (321,3))
```

```python
a_dataframe = pd.DataFrame(reshaped_array, columns=["Year", "Song", "Artist"])
```

```python
a_dataframe
```

```python
# Dropping 

a_dataframe['Song'] = a_dataframe['Song'].apply(lambda x: (str(x).replace('"', '')))
```

```python
a_dataframe # anti-war songs from wiki!
```

```python
a_dataframe = a_dataframe.drop(['Year'], axis=1)
a_dataframe
```
